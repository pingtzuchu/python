```python
import xml.etree.ElementTree as ET
import re
tree = ET.parse("/Users/pingtzuhu/Google 雲端硬碟/Cloud Drive/程式資料/Oxygen/SHY/shy-1-TEI-P5.xml")
root = tree.getroot()
ns = {"tei":"http://www.tei-c.org/ns/1.0"}
firstString=""
secondString=""
listOfHeading=[]
listOfcontent=[]
tempDiv=ET.fromstring("<div></div>")
tempFirstDiv=ET.fromstring("<div></div>")
tempSecondDiv=ET.fromstring("<div></div>")
body = root.find("./tei:text/tei:body", ns)
currentNode=body
collectFirstString=False
contentInDiv=False
divLevel=0
repetition=0
'''
judge whether div or not
if yes, adding a div and set it as currentNode
if no, append it to currentNode
delete the p
if div having the same page number, turn them into hierarchical structure
'''
for p in body.findall("tei:p", ns):
    if p.get("rend")=="heading 3" and p.text.find("(P")> -1:
        body.append(ET.fromstring("<div>"+p.text+"</div>"))
    else:
        if divLevel==0:
            body.append(ET.fromstring("<div></div>"))
        else:
            body.find("tei:div[last()]", ns).append(p)
    body.remove(p)
tree.write("/Users/pingtzuhu/Google 雲端硬碟/Cloud Drive/程式資料/Oxygen/SHY/shy-new1.xml", encoding="UTF-8")
```
