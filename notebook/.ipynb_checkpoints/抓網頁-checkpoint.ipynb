```python
import requests
import re
from bs4 import BeautifulSoup
import lxml
def getBook(title):
    site = 'https://zh.wikisource.org'
    res = requests.get('https://zh.wikisource.org/wiki/'+title)
    soup = BeautifulSoup(res.text, "lxml")
    newFile="""<?xml version="1.0" encoding="UTF-8"?>
    <?xml-model href="http://www.tei-c.org/release/xml/tei/custom/schema/relaxng/tei_all.rng" type="application/xml" schematypens="http://relaxng.org/ns/structure/1.0"?>
    <?xml-model href="http://www.tei-c.org/release/xml/tei/custom/schema/relaxng/tei_all.rng" type="application/xml"
        schematypens="http://purl.oclc.org/dsdl/schematron"?>
    <TEI xmlns="http://www.tei-c.org/ns/1.0">
      <teiHeader>
          <fileDesc>
             <titleStmt>
                <title>Title</title>
             </titleStmt>
             <publicationStmt>
                <p>Publication Information</p>
             </publicationStmt>
             <sourceDesc>
                <p>Information about the source</p>
             </sourceDesc>
          </fileDesc>
      </teiHeader>
      <text>
          <body>"""
    for link in soup.find_all("a", title=re.compile(title+"/")):
        href = link["href"]
        title = link["title"][3:]
        page = BeautifulSoup(requests.get(site+href, "lxml").text)
        newFile += "<div><head>"+title+"</head>"
        for p in page.find_all("p"):
            newFile += p.prettify()
        newFile += "</div>"
    newFile += "</body></text></TEI>"
    with open("d:\\temp\\"+title+".xml", "w", encoding="utf-8") as outFile:
        outFile.write(newFile)
getBook("title")
print("We're done.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re, time
from bs4 import BeautifulSoup
driver = webdriver.Firefox(executable_path=r"D:\drivers\geckodriver.exe")
firstPage=driver.get("http://taco.ith.sinica.edu.tw/tdk/index.php?title=特殊:用戶登錄")
driver.find_element_by_id("wpName1").send_keys("pingtzuchu")
password = driver.find_element_by_id("wpPassword1").send_keys("kaijun2")
elem = driver.find_element_by_id("wpRemember")
time.sleep(5)
elem.send_keys(Keys.RETURN)
time.sleep(10)
base_uri ="http://taco.ith.sinica.edu.tw"
diary="http://taco.ith.sinica.edu.tw/tdk/水竹居主人日記"
newFile="<人>"
driver.get(diary)
WebDriverWait(driver, 300).until(
    EC.presence_of_element_located((By.ID, "mw-content-text"))
)
pageSource=driver.page_source
soup = BeautifulSoup(pageSource).find(id="mw-content-text")
newFile += "<page>"+soup.prettify()+"</page>\n"
fileNum = 1
controlNum = 15
for yearLink in soup.find_all("a", title=re.compile("水竹居主人日記/")):
    if fileNum > controlNum:
        href = yearLink["href"]
        newFile += "<年><page><title>"+yearLink["title"]+"</title>"
        driver.get(base_uri+href)
        WebDriverWait(driver, 300).until(
            EC.presence_of_element_located((By.ID, "mw-content-text"))
        )
        pageSource=driver.page_source
        yearsoup = BeautifulSoup(pageSource).find(id="mw-content-text")
        newFile += yearsoup.prettify()+"</page>\n"
        for monthLink in yearsoup.find_all("a", title=re.compile("水竹居主人日記/")):
            href = monthLink['href']
            newFile += "<月><page><title>"+monthLink["title"]+"</title>"
            driver.get(base_uri+href)
            WebDriverWait(driver, 300).until(
                EC.presence_of_element_located((By.ID, "mw-content-text"))
            )
            pageSource=driver.page_source
            monthsoup=BeautifulSoup(pageSource).find(id="mw-content-text")
            newFile += monthsoup.prettify()+"</page>\n"
            for dayLink in monthsoup.find_all("a", title=re.compile("水竹居主人日記/")):
                href = dayLink["href"]
                newFile += "<日><page><title>"+dayLink["title"]+"</title>"
                driver.get(base_uri+href)
                WebDriverWait(driver, 300).until(
                    EC.presence_of_element_located((By.ID, "mw-content-text"))
                )
                pageSource=driver.page_source
                dayPage=BeautifulSoup(pageSource).find(id="mw-content-text")
                newFile += dayPage.prettify()+"</page>\n</日>"
            newFile += "</月>"
        newFile += "</年>"
        with open("d:\\temp\\水竹居主人日記_"+"0"*(2-len(str(fileNum)))+str(fileNum)+".xml", "w", encoding="utf8") as outFile:
            outFile.write(newFile+"</人>")
        newFile="<人>"
    print(fileNum)
    fileNum += 1
print("We're done.")
    
```
