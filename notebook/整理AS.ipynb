```python
import re, os
from bs4 import BeautifulSoup
filepath = r"G:\小組雲端硬碟\計畫控制\shy"
savePath = r"D:\AS"
headLengthToCut = int(input("input head lenght to Cut: "))
startFileNum = int(input("input staring file number: "))
endFileNum = int(input("input ending file number: "))
missingFiles=[]
initial = 1 
savedFileNumber = 6
for fileNum in range(startFileNum, endFileNum+1):
    #檔案存在就開啟
    try: 
        infile = open(filepath+"\\"+str(fileNum).strip()+".html", "r",
             encoding="utf8")
        IsFileExist=True
    except:
        missingFiles.append(fileNum)
        IsFileExist=False
    #檔案不存在就跳到下一個檔案號碼
    if IsFileExist: 
        print(fileNum)
        soup = BeautifulSoup(infile.read(), "xml")
        theHead = str(soup.head.string)
        headLen = len(theHead.split("／"))
        theSpan = soup.find("span", id="fontstyle")
        try:
            theBegin = str(theSpan.div.string)[:5]
        except:
            theBegin = ""
    #如果是起始狀態，設定好起始值後離開；如果該存檔案就存檔
        if (initial == 1) or (headLen == headLengthToCut):
            headPrev = theHead
            spanPrev = theSpan
            beginPrev = theBegin
            finalXML = BeautifulSoup("<div></div>", "xml")
            currentDiv = finalXML.div
            newDiv = finalXML.new_tag("div")
            if initial == 1:
                for i in range(headLen-headLengthToCut-1):
                    newDiv = finalXML.new_tag("div")
                    currentDiv.append(newDiv)
                    currentDiv = currentDiv.div
            else:
                currentDiv.append(spanPrev)
                savedFileNumberString=str(savedFileNumber).strip()
                with open(savePath+"\\" + "0"*(4-len(savedFileNumberString))+
                          savedFileNumberString+".xml",
                          "w", encoding="utf8") as outfile:
                    outfile.write(finalXML.prettify())
                    savedFileNumber += 1
                    print("file: "+savedFileNumberString)
            newHead = finalXML.new_tag("head")
            currentDiv.append(newHead)
            newHead.string=theHead.split("／")[-1]
            initial = 0
        else:
            if theHead == headPrev: #如果兩頁標題一樣，跳到下一頁
                pass
            else:
                if headPrev in theHead: #如果前頁標題包含在下一頁
                    #檢查頁前5字是否不一樣，不一樣的話也加入內容
                    if not theBegin == beginPrev: 
                        currentDiv.append(spanPrev)
                else: #不包含在下一頁，儲存舊的標題內容，移動Level Level只會往上升
                    currentDiv.append(spanPrev)
                    for i in range(len(headPrev.split("／"))-headLen+1):
                        currentDiv = currentDiv.parent
                spanPrev = theSpan
                headTag = finalXML.new_tag("head")
                newDiv = finalXML.new_tag("div")
                currentDiv.append(newDiv)
                currentDiv = newDiv
                currentDiv.append(headTag)
                headTag.string = theHead
                headPrev = theHead
                beginPrev = theBegin
                if fileNum==endFileNum:
                    try:
                        currentDiv.append(spanPrev)
                    except:
                        pass
                    savedFileNumberString=str(savedFileNumber).strip()
                    with open(savePath+"\\" + "0"*(4-len(savedFileNumberString))
                              + savedFileNumberString+".xml",
                              "w", encoding="utf8") as outfile:
                        outfile.write(finalXML.prettify())
                        print("file "+savedFileNumberString+" saved")
print(missingFiles)
```
